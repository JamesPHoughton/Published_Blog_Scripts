This research has several goals:

- Understand the ecosystem of APIs, mashups/Apps, and end users, and the drivers and delays in each of the interactions between these actors
- Understand what makes some APIs 'stick' while others fade

The folks at [Mulesoft](http://www.mulesoft.com/) were kind enough to provide an export of the Programmable Web Mashups table, including a list of mashups, the APIs that they use, the date they were added to programmable web, and the tags that help identify and categorize them.

Before we try to apply any models or inference algorithms, lets see how much we can learn about the API ecosystem purely by looking at this data.

<!-- If we ever publish this notebook and the data file anywhere, it might make sense to parse the lists differently, or save the data frame as a pickle? -->

```python
import pandas as pd
mashups = pd.read_csv('mashups-direct.csv', header=None, names=['name','date','APIs','tags'], 
                      index_col='date', parse_dates=True)

# Split the 'string' API and Tag lists into python sets
mashups['API List'] = mashups['APIs'].apply(lambda x: set(str(x).split(',')))
mashups['Tags List'] = mashups['tags'].apply(lambda x: set(str(x).split(',')))

mashups.head(2)
```

### Mashups Added Per Month

Lets first take a look at what data is available to us chronologically, by plotting new additions to the PW database vs time.

It's interesting to observe that the rate at which mashups have been added to the dataset has declined. Before viewing the data, my expectation would have been that more mashups are being started in 2013-2014 than were being started in 2007. This may show either that the market space for mashups is being saturated, or that the fraction of mashups cataloged in Programmable Web has declined. 

We would likely need an external datasource to determine the degree to which each is true. If [this unverifiable website](http://148apps.biz/app-store-metrics/?mpage=submission) is to be believed, the number of submissions to the Apple App Store per month has increased over the same timeperiod. While this doesn't represent Mashups directly, we can assume that the two are likely correlated, which suggests that the decline in listings of new mashups in programmable web may not be representative of actual behavior.

Going forward **it will be important for us to verify the representativeness of the PW data over time**.

```python
# First we get a count of the monthly additions to PW
mashup_monthly = pd.DataFrame()
mashup_monthly['Additions Each Month'] = mashups['name'].resample(rule='M', how='count')

# Then we calculate a running sum
mashup_monthly['Running Total'] = mashup_monthly['Additions Each Month'].cumsum()

# Show the result
fig, ax1 = plt.subplots(figsize=(12,5))
ax2 = ax1.twinx()

ax1.bar(mashup_monthly.index, mashup_monthly['Additions Each Month'], width=20, color='b', alpha=.5, linewidth=0)
ax1.set_ylabel('Additions each Month', color='b', alpha=.5, fontsize=18)

ax2.plot(mashup_monthly.index, mashup_monthly['Running Total'], 'r-', linewidth=2)
ax2.set_ylabel('Running Total', color='r', fontsize=18)
plt.title('All Mashups Listed in Programmable Web', fontsize=18);
```

### APIs per Mashup, over time

Lets also look at how the total number of APIs referenced in each mashup changes over time. For this chart to make sense, we'll aggregate to the monthly level, and scatterplot datpoints, understanding that a true 'average rate' will be estimated by the general trend.

Before looking at the data, we expect the number of APIs used in each mashup to increase over time, as the total number of APIs increases, and as developers have more time to integrate them. This seems to be the case for a while; although we may see some leveling off more recently. The datapoints are more sparse here, and so the variance between months increases.

```python
# Count the number of mashups in each api
mashups['Num APIs'] = mashups['API List'].apply(len)

# Calculate the monthly average number of mashups per api
mashup_monthly['Num APIs'] = mashups['Num APIs'].resample(rule='M', how='mean')

# Display the result
plt.figure(figsize(8,5))
plt.plot(mashup_monthly.index, mashup_monthly['Num APIs'], 'o', alpha=.75, linewidth=0)
plt.ylabel('APIs per Mashup', fontsize=18)
plt.title('Monthly Average Number of APIs per Mashup', fontsize=18);
```

### Most Popular APIs

Lets have a look and see which APIs in the dataset have the most number of mashups calling on them. We see that the usual suspects - the large internet players - have good representation, although the API presence of each of the players probably doesn't align exactly with their internet footprints, meaning that some have a larger fractional dependence on APIs than others do.

```python
#Count the uses of each API. We chain the values in the api list together, and count that full list.
import itertools
API_counts = pd.Series(list(itertools.chain(*mashups['API List'].values))).value_counts()

API_counts.iloc[:20].plot(kind='barh', figsize=(8,5), linewidth=0);
plt.title('Most Popular APIs', fontsize=18);
```

Its unsurprising that some APIs tend to dominate the results. The shape of this curve suggests a power law distribution. If we plot this on a log-log plot, we should see something like a straight line, and we do.

### Most popular API sets

In addition to looking at the most popular individual APIs, we can look at the most popular common 'sets' of APIs. This shows both mashups with multiple and single APIs listed. Because there are so many mashups that use just one api, we have to go fairly far down the list before we come across sets with multiple APIs listed (here in red).

```python
combo_counts = mashups['APIs'].value_counts()
combo_counts.iloc[:20].plot(kind='barh', figsize=(8,5), linewidth=0,
                            colors=['r' if ',' in x else 'b' for x in combo_counts.iloc[:20].index])
plt.title('Most Popular API Combinations', fontsize=18)
```

To me, this paucity of multiple APIs in each mashup says that the most popular way to 'Mash Up' the most popular APIs is to just provide a user interface for a single service. We can show this by counting the number of mashups that have 1, 2, 3... different APIs.

```python
APIs_per_mashup = mashups['API List'].apply(len).value_counts().sort_index()
APIs_per_mashup.plot(kind='bar')
plt.xlabel('Number of APIs per Mashup', fontsize=15)
plt.ylabel('Number of Mashups', fontsize=15);
plt.title('Number of APIs per Mashup, by Count')
```

Over half of the mashups listed on PW seem to be just front ends for the API, and don't combine multiple services.

```python
plt.pie(APIs_per_mashup, labels=APIs_per_mashup.index, labeldistance=.9);
plt.axis('equal');
plt.title('Number of APIs per Mashup, by Frequency');
```

### API comparisons
Next we'll make some comparisons between APIs that are in competition with one another. This helper function plots combinations of interesting APIs against one another.

```python
def API_compare(interest_list, category_name, normed=False):
    
    interest_mashups = mashups['API List'].apply(lambda x: ', '.join(x.intersection(interest_list)))
    interest_mashups = interest_mashups[interest_mashups != ''] #drop empty rows
    
    groups = interest_mashups.groupby(interest_mashups).groups
    
    interest_timecount = pd.DataFrame(index=interest_mashups.resample(rule='Q', how='count').index)
    for groupname in groups:
        interest_timecount[groupname] = pd.Series(index=groups[groupname], data=1).resample(rule='Q', how='count')
    
    interest_timecount.fillna(0, inplace=True) 
    
    if normed:
        interest_timecount = interest_timecount.div(interest_timecount.sum(axis=1), axis=0)
        
    # Display a stack of these values
    plt.figure(figsize=(8,5))
    stack_coll = plt.stackplot(interest_timecount.index, interest_timecount.values.T, 
                               linewidth=0, colors=contrastcolors)
    
    # make the legend
    proxy_rects = [Rectangle((0, 0), 1, 1, fc=pc.get_facecolor()[0], linewidth=0) for pc in stack_coll]
    plt.title('New Invokations of %s each Quarter%s'%(category_name, ', Normalized' if normed else ' '), fontsize=18)
    plt.legend(proxy_rects, list(interest_timecount.columns.values), loc='upper left', bbox_to_anchor=(1,1)); 
        
    if normed:
        plt.ylim(0,1)
        
```

#### Competition between Map APIs

First lets look at the primary offerings for maps. Looking at the data on a quarterly basis, we see that in addition to decline in new invocations of each API, there is consolidation to the leader - possibly counterintuitive behavior; and any mashups involving multiple mapping APIs tending to replace new single offerings offerings - probably more intuitive.

```python
API_compare({'Google Maps', 'Microsoft Bing Maps', 'Yahoo Maps'}, category_name='Map APIs')
```

```python
API_compare({'Google Maps', 'Microsoft Bing Maps', 'Yahoo Maps'}, category_name='Map APIs', normed=True)
```

#### Geocoding APIs

Here we're comparing essentially commodity products - all of these services take an address and convert it into a set of latitude and longitude coordinates. These APIs are competing almost exclusively on ease of use.

We see that despite being a late entry, Google's geocoder represents the largest growth component toward the end of the timeseries.

```python
API_compare({'Yahoo Geocoding', 'geocoder', 'Google Geocoding', 'geocoder.ca', 'Bing Maps Geocode'}, 
            category_name='Geocoding APIs')
```

When we look at the normalized version of this chart, we're seeing essentially the 'new app market share' - here it seems that when Google came on the scene Yahoo ceded this 'new development market share' strongly to Google.

```python
API_compare({'Yahoo Geocoding', 'geocoder', 'Google Geocoding', 'geocoder.ca', 'Bing Maps Geocode'}, 
            category_name='Geocoding APIs', normed=True)
```

What we can't really plot here is the actual market share. The true market share depends not only on entrants to the market, but also the number of mashups that retire, over time.
